# Imports

```python
from fastai.vision.all import *
```

```python
path = Path('clothing')
```

# Load the Data using the DataBlock API

```python
clothing = DataBlock(
    #types for the independent and dependent variables
    blocks=(ImageBlock, CategoryBlock), 
    #returns a list of all the images in the path (it takes a path as argument)
    get_items=get_image_files, 
    #how to split the dataset: 20% validation set, 80% training set. Split randomly
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    #how to create the labels for the dependent variable
    get_y=parent_label,
    # resizing the images to the same size
    item_tfms=Resize(224, ResizeMethod.Crop),
    #adding data augmentation
    batch_tfms=aug_transforms())
```

```python
#A DataLoaders includes validation and training DataLoaders. 
#DataLoader is a class that provides batches of a few items at a time to the GPU (64 items).
dls = clothing.dataloaders(path)
```

# Showing a few items of our Dataset

```python
dls.train.show_batch(max_n=4, nrows=1)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

# Checking Data Augmentation technique used

```python
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

# Fitting Data to our model

```python
#Defining a learner
learn=cnn_learner(dls,resnet34,metrics=error_rate)
```

```python
# Using the learning rate finder to find the best learning rate for updating the weights of the last layers.
learn.lr_find()
```

```python
lr=8e-3
learn.fit_one_cycle(n_epoch=6,lr_max=lr)
```

```python
# unfreezing the model to enable gradient descent to update the weights of the last layers
learn.unfreeze()
```

```python
# Using the learning rate finder to find the best learning rate for updating the weights of the whole CNN.
learn.lr_find()
```

```python
learn.fit_one_cycle(n_epoch=2,lr_max=1e-5)
```

# Looking the mistakes of the model

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

# Saving the model

```python
learn.export()
```
